#Matplotlib and Seaborn


```python
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
x=np.arange(40, 50)
y=np.arange(50, 60)
```

```python
min_len = min(len(x), len(y))
plt.scatter(x[:min_len], y[:min_len])
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')
plt.grid(True)
plt.savefig('scatter_plot.png')
# plt.show()
```

```python
plt.plot(x, y, color='red', linestyle='dashed', linewidth=2, marker='*', markerfacecolor='red', markersize=10)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot')
plt.grid(True)
```

```python
plt.subplot(2, 2, 1)
plt.plot(x, y, color='red', linestyle='dashed', linewidth=2, marker='*', markerfacecolor='red', markersize=10)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot')
plt.grid(True)

plt.subplot(2, 2, 2)
plt.scatter(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')
plt.grid(True)

plt.subplot(2, 2, 3)
plt.bar(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Bar Plot')
plt.grid(True)

plt.subplot(2, 2, 4)
plt.hist(x)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Histogram')
plt.grid(True)

plt.tight_layout()
plt.savefig('subplot_plot.png')
# plt.show
```

```python
#Histogram

plt.subplot(2, 2, 1)
plt.hist(x, bins=5, color='green', edgecolor='black', alpha=0.7)
plt.xlabel('X-axis Values')
plt.ylabel('Frequency')
plt.title('Histogram of X')
plt.grid(True)

#boxplot

plt.subplot(2, 2, 2)
plt.boxplot(x, notch=True, vert=True, patch_artist=True,
            meanline=True, showmeans=True,
            boxprops=dict(facecolor='lightblue', color='blue'),
            medianprops=dict(color='red'),
            meanprops=dict(marker='o', markeredgecolor='black', markerfacecolor='green'),
            flierprops=dict(marker='x', markersize=8, markeredgecolor='orange'),
            whiskerprops=dict(color='purple'))
plt.xlabel('')
plt.ylabel('Values')
plt.title('Box Plot of X')
plt.grid(True)

#Pie

plt.subplot(2, 2, 3)
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)  # only "explode" the 2nd slice (i.e. 'B')
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Pie Chart Distribution')
plt.grid(True)

#scatter

plt.subplot(2, 2, 4)
plt.scatter(x, y, color='red', marker='o', s=100, alpha=0.7)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')
plt.grid(True)

plt.tight_layout()
plt.savefig('subplot_plot_max.png')
# plt.show
```
